<pre>
C/A,UNIT,SCP,STATION,LINENAME,DIVISION,DATE,TIME,DESC,ENTRIES,EXITS
  
C/A      = Agrupamento de catracas de que a catraca faz parte (_Control Area_)
UNIT     = Cabine de controle associada à estação onde a catraca se encontra (_Remote Unit for a station_)

LINENAME = Código representando todas linhas que passam na estação*
DIVISION = Código representando a concessionária original da linha, antes da prefeitura assumir a gestão   
DATE     = Representa a data (no formato MM-DD-YY) do evento de auditoria agendado
TIME     = Representa o horário (hh:mm:ss) do evento de auditoria agendado
DESc     = Descreve o tipo de evento de auditoria registrado:
           1. "REGULAR" representando um evento de auditoria padrão, em que a contagem é feita a cada 4 horas
           2. "RECOVR AUD" significa que o valor específico estava perdido, mas foi recuperado posteriormente 
           3. Diversos códigos sinalizam situações em que auditorias são mais frequentes devido a atividades de
              planejamento ou solução de problemas. 
ENTRIES  = A contagem cumulativa de entradas associadas à catraca desde o último registro
EXITS    = A contagem cumulativa de saídas associadas à catraca desde o último registro

*  Normalmente as linhas são representadas por um caractere. LINENAME 456NQR significa que os trens 4, 5, 6, N, Q e R passam pela estação.

````python
import os
import urllib
import re
import pandas as pd
import collections
from bs4 import BeautifulSoup
from datetime import datetime
%matplotlib notebook
````

````python
def load_dataset(sufix_url, base_url = "http://web.mta.info/developers/"):
    """
    Função utilizada para baixar os arquivos de dados do metrô de New York.
      Recebe a url base e o final da url, pois assim temos uma forma mais genérica de trabalhar
      com os dados. Caso mude a url, basta passar via parâmetro.
      ----------------------------------------------------------
      sufix_url: final da url, o link do documento.
      base_url: a url do site, no caso o default é http://web.mta.info/developers/
    """
    
    u = urllib.urlopen(base_url+sufix_url)
    html = u.read()
    soup = BeautifulSoup(html, "html.parser")
    links = soup.find_all('a')
    for link in links:
        try:
            if link and '1706' in link.get('href'):
                filename = link.get('href').split('/')
                filename = filename[len(filename) - 1]
                urllib.urlretrieve(base_url + link.get('href'), filename)
                print("Lendo arquivo de {}. \nSalvando arquivo como {}"
                      .format(base_url + link.get('href'), filename))
        except:
            #Caso ocorra uma excessão lendo o arquivo do link(ex: 404), ignora e continua para o próximo
            continue
    
````

### *Exercicio 1.2*

Escreva uma função que pegue a lista de nomes dos arquivos que você baixou no exercicio 1.1 e consolide-os em um único arquivo. Deve existir apenas uma linha de cabeçalho no arquivo de saida. 

Por exemplo, se o arquivo_1 tiver:
linha 1...

e o outro arquivo, arquivo_2 tiver:
linha 3...
linha 4...

Devemos combinar o arquivo_1 com arquivo_2 em um arquivo mestre conforme abaixo:

'C/A, UNIT, SCP, DATEn, TIMEn, DESCn, ENTRIESn, EXITSn'
linha 1...
linha 2...
linha 3...
linha 4...
linha 5...

**OBS:** Note que algumas colunas foram descartadas!

````python
def create_master_turnstile_file(filenames, output_file):
    """
        Usado para unificar mais de um arquivo em um único, pula a primeira linha de cada arquivo.
        Recebe uma lista de nomes de arquivos e o nome do arquivo de saída. 
        -------------------------------------------------------------------
        filenames: array de nomes de arquivos.
        output_file: o nome do arquivo de saída, onde vão ser salvo os dados dos arquivos de entrada.
    """
    with open(output_file, 'w') as master_file:
        master_file.write('C/A,UNIT,SCP,DATEn,TIMEn,DESCn,ENTRIESn,EXITSn\n')
        for filename in filenames:
            f = open(filename,"r")
            if f.mode == 'r':
                data = f.readlines()[1:]
                for line in data:
                    master_file.write(line)
                f.close()
````

````python
def get_file_names(path = "."):
    """
        Usado para carregar uma lista de arquivos, todos devem conter no nome o valor 1706.
        TODO: usar novo parâmetro >> default_name = '1706'
        -------------------------------------------------------------------
        path: diretório de onde será lido os arquivos, default é o diretório 
              de onde se encontra rodando o programa.
    """
    files = []
    listOfFiles = os.listdir(path) 
    for file in listOfFiles:
        if '1706' in file:
            files.append(file)
    return(files)
````

````python
create_master_turnstile_file(get_file_names(),"master.txt")
````

````python
def get_rows_difference(df, column, new_column):
    """
        Função utilizada para pegar a diferença entre a linha posterior e linha anterior
        de um dado Dataframe.
        -------------------------------------------------------------------
        df: o Dataframe com os dados a serem analisados.
        column: coluna que será calculada.
        new_column: o nome da nova coluna, gerada a partir do cálculo da coluna informada.
    """
    aux = df.copy()
    aux[new_column] = aux[column].shift()
    aux[new_column] = aux[column] - aux[new_column]
    aux[new_column].fillna(1, inplace=True)
    return aux
````

### *Exercicio 1.6*

Dado uma variável de entrada que representa o tempo no formato de:
     "00:00:00" (hora: minutos: segundos)
    
Escreva uma função para extrair a parte da hora do tempo variável de entrada
E devolva-o como um número inteiro. Por exemplo:
         
         1) se a hora for 00, seu código deve retornar 0
         2) se a hora for 01, seu código deve retornar 1
         3) se a hora for 21, seu código deve retornar 21
        
Por favor, devolva a hora como um número inteiro.


````python
def time_to_hour(time, pattern = '\d{2}:\d{2}:\d{2}', extract = '\d{2}'):
    """
        Função utilizada para pegar converter uma String em formato de tempo para hora.
        -------------------------------------------------------------------
        time: a string a ser decomposta.
        pattern: a regex da string informada. Default:\d{2}:\d{2}:\d{2}
        extract: a regex a ser extraída da string. Default: \d{2}
    """
    
    result = re.match(pattern, time)   
    if result and result.group(): 
        aux = re.search(extract, time)
        if aux.group():
            return(int(aux.group(0)))
    raise Exception('String {} don\'t have the same pattern as {} or {} couldn\'t be found.'
                    .format(time, pattern, extract))
````

Exporte seu mapeador em um arquivo chamado mapper_result.txt e envie esse arquivo juntamente com a sua submissão. O código para exportar seu mapeador já está escrito no código abaixo.




````python
import sys
import codecs
sys.stdout = codecs.getwriter('utf8')(sys.stdout)

def mapper():
    first_line = True
    for line in sys.stdin:  
        line = line.decode('iso8859-1')
        if first_line:
            first_line = False
            continue
        aux = line.split(",")
        if len(aux) > 6:
            print("{}\t{}".format(aux[1], aux[6]).encode('utf-8'))  
        

sys.stdout = open('mapper_result.txt', 'w')
mapper()
sys.stdin = open('turnstile_data_master_with_weather.csv')
````

Exporte seu redutor em um arquivo chamado reducer_result.txt e envie esse arquivo juntamente com a sua submissão.

````python
def reducer():
    table = dict()

    for line in sys.stdin:
        aux = line.split('\t')
        if table.get(aux[0], 0) > 0:
            table[aux[0]] = table.get(aux[0], 0) + float(aux[1])
        elif len(aux) == 2:
            table[aux[0]] = float(aux[1])
    table = collections.OrderedDict(sorted(table.items()))
    for key in table:
        print("{}\t{}".format(key, table[key]).encode('utf-8'))  

sys.stdout = open('reducer_result.txt', 'w')
reducer()
sys.stdin = open('mapper_result.txt')
````
